```python
def get_args(args, relevance):
    new_args = []
    for row in args:
        temp = []
        for arg in row:
            if arg['relevance'] in relevance:
                temp.append(arg['argument'])
        new_args.append(temp)
    
    return new_args
```

```python
from rapidfuzz.distance.LCSseq import normalized_similarity
from itertools import combinations

def check_similarity(arguments_list):
    if arguments_list == [None]:
        return arguments_list
    else:
        removed = set()
        result = []
        for argument in arguments_list:
            if argument not in removed:
                result.append(argument)
                similar = [pair[1] for pair in combinations(arguments_list, 2) if pair[0] == argument and normalized_similarity(pair[0], pair[1]) > 0.8]
                removed.update(similar)
        return result
```

```python
new_args = [check_similarity(args) for args in new_args]
new_args[:5]
```
